```python
extraction=list(df.loc[:,'Hyperlink'])
extraction
```

```python
cars_n_dataset={'model_name':[]}
```

```python
for url in extraction:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    model_name = soup.find('h1', class_="model-overview-title font-muli-bold").text
    cars_n_dataset['model_name'].append(model_name)
```

```python
cars_dataset=[]
for url in extraction:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the table containing the desired information
    table = soup.find('table', class_='keySpecsTable')
    if table:
        # Initialize dictionary to store data for this car
        car_data = {}
        
        # Loop through each row in the table
        for row in table.find_all('tr'):
            # Extract label and value from each row
            label = row.find('th').text.strip()
            value = row.find('td').text.strip()
            
            # Store data in the car_data dictionary
            car_data[label] = value
        
        # Append car_data to cars_dataset
        cars_dataset.append(car_data)

```

```python
df_cars=pd.DataFrame(cars_dataset)
```

```python
df_cars.insert(loc=0,column='Name',value=cars_n_dataset['model_name'])
```

```python
search_urls = []
for model in df_cars['Name']:
    stripped_model = model.replace(' ', '%20')  # Replace spaces with %20 for URL encoding
    url = 'https://auto.hindustantimes.com/search?text=' + stripped_model
    search_urls.append(url)

print(search_urls)
```

```python
hlinks=[]
for searches in search_urls:
    response = requests.get(searches)
    soup=BeautifulSoup(response.text,'html.parser')
    finding_hlink=soup.find('a',class_='parentWrapper')
    href=finding_hlink.get('href')
    hlinks.append(href)
```

```python
hlinks
```

```python
# Construct the search query
top_speeds=[]
for url in hlinks:

    # Send request to the URL
    response = requests.get(url)

        # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

        # Find the table containing key specifications
    specs_table = soup.find('table', class_='KeySpecs_KeySpecsSectionWrapper__F_3hc')

        # Check if the specs table is found
    if specs_table:
            # Find all rows in the table
        rows = specs_table.find_all('tr')

            # Loop through each row to find the top speed
        for row in rows:
                # Check if the row contains the key 'Top Speed'
            if 'Max Speed' in row.text:
                    # Extract the top speed value from the next row
                top_speed_element = row.find_next('td', class_='KeySpecs_KeySpecsSubtitle__X0vwC')
                top_speed = top_speed_element.get_text(strip=True)
                top_speeds.append(top_speed)
```
